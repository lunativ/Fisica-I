Ejercicio 4: Una pelota se deja caer desde una altura de 2m y rebota hasta una altura de 1,5m. 

Calcular la fuerza neta que ejerce la pelota al chocar con el piso si el tiempo de impacto fue de $10^{-2}s$:

Para calcular la fuerza neta debemos primero calcular el impulso mediante la relacion: $I = P_f - P_i = m * v_f - m * v_i$. Para esto necesitamos conocer las velocidades inmediatamente antes y despues del choque, aplicando las ecuaciones de caida libre durante el tramo de caida y tiro vertical en la subida.

1) Primero calculamos la velocidad justo antes del choque, considerando que la pelota cae con movimiento de caida libre desde la altura inicial de 2m hasta el piso:

$$y_i = 2m, v_1i = 0, g = 9.8 \frac{m}{s^2}$$ 
$$ {v_{1f}}^2 = {v_{1i}}^2 + 2*g*(y_f - y_i )$$
Despejando $${v_{1f}} = -6.28 \frac{m}{s}$$

Además, podemos calcular el tiempo de caida resolviendo: 

$$y_{1f} = y_{1i} + v_{1i}  t + \frac{1}{2}  g  t^2$$
Resolviendo las raices: 
$$t = 0,639s$$

2) Calculamos la velocidad justo despues del choque considerando al tramo de subida de la pelota como un movimiento de tiro vertical con velocidad inicial igual a la velocidad despues del choque y velocidad final igual a 0.

$$y_{2i} = 0m, y_{2f} = 1,5m v_{2f} = 0, g = 9.8 \frac{m}{s^2}$$ 
$$ {v_{2f}}^2 = {v_{2i}}^2 + 2*g*(y_{2f} - y_{2i} )$$
Despejando $${v_{2i}} = 5.42 \frac{m}{s}$$

De la misma manera que en el punto anterior, podemos calcular el tiempo de subida mediante: 
$$y_{2f} = y_{2i} + v_{2i}  t + \frac{1}{2}  g  t^2$$
Resolviendo las raices: 
$$t = 0,55s$$
3) Ahora podemos calcular el impulso I:

$$I = P_f - P_i = m * v_{1f} - m * v_{2i}$$

$$I = 0,1 Kg (5.42 - (-6.28)) \frac{m}{s} = - 1,17 Kg\frac{m}{s}$$

$$F_N = \frac{I}{t} = \frac{-1,17 Kg\frac{m}{s}}{0.01s} = 117 N$$


A continuación se grafica la trayectoria de la pelota, su velocidad y energía mecánica: 

```python
from matplotlib import *
from numpy import *
from pylab import *
y1 = linspace(2, 0, 1000)
g = -9.8
h0 = 2

t1 = linspace(0,0.639,50)
v1 = g*t1
y1 = h0 + 0.5*g*t1**2

t2 = linspace(0,0.55,50)
v2i = 5.42
v2 = v2i + g*t2
y2 = v2i*t2 + 0.5*g*t2**2


plt.figure(1)
plot(t1,y1,'r',label='y1')
plot(t2+0.639,y2,label='y2')
plt.xlabel('Tiempo (s)')
plt.ylabel('Altura (m)')
grid()
plt.legend(loc='upper right')
plt.title('Altura de la pelota en función del tiempo')
show()
plt.figure(2)
plot(t1,v1,'r',label='v1')
plot(t2+0.639,v2,label='v2')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
grid()
plt.legend(loc='upper right')
plt.title('Velocidad de la pelota en función del tiempo')
show()

#Energia cinetica
m = 0.1
k1 = 0.5 * m * v1**2
k2 = 0.5 * m * v2**2

#Energia potencial gravitatoria
g = 9.8
Ug1 = m * g * y1 
Ug2 = m * g * y2 

plt.figure(3)
plot(t1,k1,'r',label='K1')
plot(t2+0.638,k2,label='K2')
plot(t1,Ug1,'m',label='Ug1')
plot(t2+0.639,Ug2,'c',label='Ug2')
plot(t1,Ug1+k1,'g',label='Em')
plot(t2+0.639,Ug2+k2,'g')
plt.xlabel('Tiempo (s)')
plt.ylabel('Energia (J)')
grid()
plt.title('Energía de la pelota en función del tiempo')
plt.legend(loc='upper right')
```
